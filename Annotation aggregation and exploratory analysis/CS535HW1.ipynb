```python
n = df2.iloc[1].sum()
N = 1470
```

```python
Pi=df2.apply(lambda num: num**2).sum(axis=1)
Pi = (Pi - n)/(n*(n-1))
print(Pi)
P_hat = Pi.mean()
P_hat
```

```python
pj=df2.sum(axis=0)/(N*n)
print(pj)
Pe = pj.apply(lambda num: num**2).sum(axis=0)
Pe
```

```python
kappa = (P_hat - Pe)/(1-Pe)
kappa
```

```python
import io
import pandas as pd
df3 = pd.read_csv(io.BytesIO(uploaded['emotion_annotation.csv']))
df3
```

```python
import scipy.stats
from scipy.stats import chi2
import numpy as np
```

```python
truth = df3['filename'].str.split(pat="_")
mylist = []
for item in truth:
    mylist.append(item[2])
pd.DataFrame(mylist)
df3['truth'] = mylist
df3
```

```python
ct_table_ind = pd.crosstab(df3['truth'], df3['HAPPY'])
print('contingency_table :\n',ct_table_ind)
```

```python
chi2_stat, p, dof, expected = scipy.stats.chi2_contingency(ct_table_ind)

print(f"chi2 statistic:     {chi2_stat:.5g}")
print(f"p-value:            {p:.5g}")
print(f"degrees of freedom: {dof}")
print("expected frequencies:\n",expected)
```

```python
ct_table_ind2 = pd.crosstab(df3['truth'], df3['SAD'])
print('contingency_table :\n',ct_table_ind2)
```

```python
chi2_stat, p, dof, expected = scipy.stats.chi2_contingency(ct_table_ind2)

print(f"chi2 statistic:     {chi2_stat:.5g}")
print(f"p-value:            {p:.5g}")
print(f"degrees of freedom: {dof}")
print("expected frequencies:\n",expected)
```
