```python
data = data[data['Race'].isin([1, 2, 3])]
data
```

```python
import numpy as np
from statistics import mean, median
```

```python
category = data.keys()[4:-1]
```

```python
def printMeanMedian(col, group, stat):
  if stat == 'mean':
    return data.groupby(group)[col].mean()
  return data.groupby(group)[col].median()
```

# Race

## Means

```python
means = pd.DataFrame()
for c in category:
  means = pd.concat([means, printMeanMedian(c, 'Race', 'mean')], axis = 1)
means
```

## Median

```python
medians = pd.DataFrame()
for c in category:
  medians = pd.concat([medians, printMeanMedian(c, 'Race', 'median')], axis = 1)
medians
```

```python
def createGroup(col, group):
  len = 7
  if(group == 'Gender'):
    len = 2
  list = []
  for i in range(len):
    list.append(data.loc[data[group] == i][col])
  return list
```

```python
dict = {}
for c in category:
  dict[c] = createGroup(c, 'Race')
```

## Bias

```python
from scipy.stats import f_oneway
```

```python
race_p_values = {}
for c in category:
  p = f_oneway(dict[c][1], dict[c][2], dict[c][3])
  race_p_values['r' + c]=p
race_p_values
```

```python
sorted_race = sorted(race_p_values.items(), key=lambda x: x[1][1])
sorted_race
```
